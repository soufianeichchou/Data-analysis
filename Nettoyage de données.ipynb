```python

file_path = "../input/chofactory/data.xlsx"
```

```python
ch = pd.read_excel(file_path, sheet_name="Dim Customers")
ch.head()
ch.rename(columns={'Fact Sales[Order ID]': 'Order_id','Fact Sales[Product ID]':
                          'Product_id','Fact Sales[Location ID]':'Location_id','Fact Sales[Sales Person ID]':
                          'Sales_Person_id','Fact Sales[Customer ID]':'Customer_id','Fact Sales[Purchase Date]':
                          'Purchase_date','Fact Sales[Quantity]':'Quantity','Fact Sales[Price]':'Price'},
                           inplace=True)
ch.head()
```

```python
modified_file_path = '/kaggle/working/NewData_Customers.xlsx'
excel_writer = pd.ExcelWriter(modified_file_path, engine='xlsxwriter')
ch.to_excel(excel_writer, sheet_name='Customers')
excel_writer.close()
```

```python
ch = pd.read_excel(modified_file_path, sheet_name="Costumers")
ch.head()
ch.drop(columns=["Order_id", "Product_id"], inplace=True)

ch.head()
```

```python
file_path = "../input/latest/NewData_Date.xlsx"
ch = pd.read_excel(file_path, sheet_name="Date")
ch = pd.read_excel(file_path, index_col=0)
ch.drop(columns=["ShortYear", "DayOfMonth","Month&Year","ShortYear"], inplace=True)
ch.head()
```

```python
modified_file_path = '/kaggle/working/NewData_Date.xlsx'
excel_writer = pd.ExcelWriter(modified_file_path, engine='xlsxwriter')
ch.to_excel(excel_writer, sheet_name='Date')
excel_writer.close()
```

```python
ch.rename(columns={'Dim Date[Date]': 'Date','Dim Date[Year]':
                          'Year','Dim Date[MonthOfYear]':'MonthOfYear','Dim Date[DayOfMonth]':
                          'DayOfMonth','Dim Date[Month&Year]':'Month&Year','Dim Date[Trimestre]':
                          'Trimestre','Dim Date[ShortYear]':'ShortYear','Dim Date[FY]':'FY',
                          'Dim Date[Month Name.1]':'Month_name','Dim Date[Day]':'Day'},
                           inplace=True)
ch.head()
```

```python
file_path = "../input/latest/NewData_Locations.xlsx"
ch = pd.read_excel(file_path, sheet_name="Location")
ch.rename(columns={'Dim Locations[Location ID]': 'Location_id','Dim Locations[Name]':
                          'Name','Dim Locations[County]':'County','Dim Locations[Nb Habitants]':
                          'Nb_habitants','Dim Locations[State Code]':'State_code','Dim Locations[State]':
                          'State','Dim Locations[Latitude]':'Latitude','Dim Locations[Longitude]':'Longitude',
                          'Dim Locations[Area Code]':'Area_code','Dim Locations[Population]':
                          'Population','Dim Locations[Households]':'Households',
                          'Dim Locations[Median Income]':
                          'Median_income','Dim Locations[Water_area]':'Water_area',
                          'Dim Locations[Column]':'Column'},
                           inplace=True)
ch = pd.read_excel(file_path, index_col=0)
ch.drop(columns=["State_code", "State","Latitude","Longitude","Area_code","Dim Locations[Water Area]"], inplace=True)
ch.head()
```

```python
modified_file_path = '/kaggle/working/NewData_Locations.xlsx'
excel_writer = pd.ExcelWriter(modified_file_path, engine='xlsxwriter')
ch.to_excel(excel_writer, sheet_name='Location')
excel_writer.close()
```

```python
file_path = "../input/chofactory/data.xlsx"
ch = pd.read_excel(file_path, sheet_name="Dim Products")
ch.rename(columns={'Dim Products[Product ID]': 'Product_id','Dim Products[Product Name]':
                          'Product_name','Dim Products[Cost]':'Cost','Dim Products[Original Sale Price]':
                          'Original_sale_price','Dim Products[Discount]':'Discount','Dim Products[Current Price]':
                          'Current_price','Dim Products[Taxes]':'Taxes'},
                           inplace=True)
ch.drop(columns=["Discount"], inplace=True)
ch.head()
```

```python
modified_file_path = '/kaggle/working/NewData_Products.xlsx'
excel_writer = pd.ExcelWriter(modified_file_path, engine='xlsxwriter')
ch.to_excel(excel_writer, sheet_name='Products')
excel_writer.close()
```
